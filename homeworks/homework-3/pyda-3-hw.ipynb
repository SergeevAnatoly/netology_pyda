## Задание 1

Дан список вида:

```python
data = [
    [13, 25, 23, 34],
    [45, 32, 44, 47],
    [12, 33, 23, 95],
    [13, 53, 34, 35]
]
```

Напишите функцию, которая возвращает сумму элементов на диагонали. Т. е. 13+32+23+35.

```python
def matrix_diagonal_sum (matrix):
    """
    Функция возвращает сумму элементов матрицы по диагонали
    """
    
    sum_result = 0
    for i, value in enumerate(matrix):
        sum_result += float(matrix[i][i])
        
    return sum_result

print(f'Сумма элементов по диагонали равна {matrix_diagonal_sum(data)}')
```

## Задание 2

Дан список чисел, часть из которых имеют строковый тип или содержат буквы. Напишите функцию, которая возвращает сумму квадратов элементов, которые могут быть числами.

```python
data = [1, '5', 'abc', 20, '2', '0']
```

```python
def string_to_float(str):
    """
    Функция преобразует входящее значение в float
    Если преобразование невозможно, возвращает 0
    """
    
    # Реализовал через try except
    try:
        if float(str):
            return float(str)
        # так как "ноль" это False, то для него ставим отдельное условие
        elif float(str) == 0:
            return float(str)
    except:
        return 0
    
def pow2(data):
    """
    Функция возвращает сумму квадратов всех элементов списка, которые могут быть преобразованы в float
    """
    return sum(\
                map(\
                    lambda x: x**2,                    #возводим в квадрат \ 
                    list(                              #получаем список результатов \
                        map(string_to_float,data)      #получаем все возможные числа из списка \
                        )\
                   )\
                )

print(f'Сумма квадратов равна {pow2(data)}')
```

Напишите программу, которая будет принимать команду от пользователя и возвращать соотвествующий результат.
Список команд: 
1 - вывести на экран среднюю оценку за все ДЗ по группе и вывести на экран среднюю оценку за экзамен по группе;
2 - вывести среднюю оценку за ДЗ и за экзамен по группе в разрезе пола студентов
3 - вывести среднюю оценку за ДЗ и за экзамен в разрезе наличия опыта в программировании у студентов.

Программа должна быть полностью декомпозирована на функции (кроме объявления функций и вызова итоговой функции в реализации ничего быть не должно).

```python
students_list = [
    {'name': 'Василий', 'surname': 'Теркин', 'sex': 'м', 'program_exp': True, 'grade': [8, 8, 9, 10, 9], 'exam': 9},
    {'name': 'Мария', 'surname': 'Павлова', 'sex': 'ж', 'program_exp': True, 'grade': [7, 8, 9, 7, 9], 'exam': 8},
    {'name': 'Ирина', 'surname': 'Андреева', 'sex': 'ж', 'program_exp': True, 'grade': [10, 9, 8, 10, 10], 'exam': 10},
    {'name': 'Татьяна', 'surname': 'Сидорова', 'sex': 'ж', 'program_exp': True, 'grade': [7, 8, 8, 9, 8], 'exam': 8},
    {'name': 'Иван', 'surname': 'Васильев', 'sex': 'ж', 'program_exp': True, 'grade': [9, 8, 9, 6, 9], 'exam': 10},
    {'name': 'Роман', 'surname': 'Золотарев', 'sex': 'м', 'program_exp': False, 'grade': [8, 9, 9, 6, 9], 'exam': 10}
]
```

```python
# Подулючаем библиотеку
import numpy as np

def avg_homework_rating(students_dict):
    """
    Возвращает среднюю оценку за все ДЗ по группе (округление до 2 десятичного знака)
    """
    # Здесь храним все оценки за ДЗ
    hw_rating = []
    
    for x in students_dict:
        for grade in x['grade']:
            hw_rating.append(int(grade))
        
    return round(np.mean(hw_rating),2)



def avg_exam_rating(students_dict):
    """
    Возвращает среднюю оценку за экзамен по группе (округление до 2 десятичного знака)
    """
    # Здесь храним все оценки за экзамены
    exam_rating = []
    
    for x in students_dict:
        exam_rating.append(int(x['exam']))
        
    return round(np.mean(exam_rating),2)



def avg_homework_rating_for_sex(students_dict):
    """
    Возвращает вывести среднюю оценку за ДЗ в разрезе пола студентов (округление до 2 десятичного знака)
    """
    # Здесь храним все оценки за ДЗ для каждого пола
    male_homework_rating = []
    female_homework_rating = []
    
    for x in students_dict:
        if x['sex'] == 'м':            
            for grade in x['grade']:
                male_homework_rating.append(int(grade))
        else:
            for grade in x['grade']:
                female_homework_rating.append(int(grade))
        
    return {'male':round(np.mean(male_homework_rating),2), 'female': round(np.mean(female_homework_rating),2)}



def avg_exam_rating_for_sex(students_dict):
    """
    Возвращает вывести среднюю оценку за экзамен в разрезе пола студентов (округление до 2 десятичного знака)
    """
    # Здесь храним все оценки за экзамены для каждого пола
    male_exam_rating = []
    female_exam_rating = []
    
    for x in students_dict:
        if x['sex'] == 'м':            
                male_exam_rating.append(int(x['exam']))
        else:
                female_exam_rating.append(int(x['exam']))
        
    return {'male': round(np.mean(male_exam_rating),2), 'female': round(np.mean(female_exam_rating),2)}



def avg_homework_rating_for_experiece(students_dict):
    """
    Возвращает вывести среднюю оценку за ДЗ в разрезе опыта программирования (округление до 2 десятичного знака)
    """
    # Здесь храним все оценки за ДЗ для тех, кто имеет и не имеет опыта программирования
    have_prog_exp_rating = []
    not_have_prog_exp_rating = []
    
    for x in students_dict:
        if x['program_exp']:            
            for grade in x['grade']:
                have_prog_exp_rating.append(int(grade))
        else:
            for grade in x['grade']:
                not_have_prog_exp_rating.append(int(grade))
        
    return {'have_exp': round(np.mean(have_prog_exp_rating),2), 'not_have_exp': round(np.mean(not_have_prog_exp_rating),2)}



def avg_exam_rating_for_experiece(students_dict):
    """
    Возвращает вывести среднюю оценку за экзамен в разрезе опыта программирования (округление до 2 десятичного знака)
    """
    # Здесь храним все оценки за экзамены для тех, кто имеет и не имеет опыта программирования
    have_prog_exp_rating = []
    not_have_prog_exprating = []
    
    for x in students_dict:
        if x['program_exp']:            
                have_prog_exp_rating.append(int(x['exam']))
        else:
                not_have_prog_exprating.append(int(x['exam']))
        
    return {'have_exp': round(np.mean(have_prog_exp_rating),2), 'not_have_exp': round(np.mean(not_have_prog_exprating),2)}



def get_result():
    # Получаем команду от пользователя (приводим сразу к int)
    command = int(input('Введите команду: '))

    if command == 1:
        print(f'Средняя оценка за все ДЗ по группе: {avg_homework_rating(students_list)}')
        print(f'Средняя оценка за экзамен по группе: {avg_exam_rating(students_list)}')
    elif command == 2:
        print(f'Средняя оценка за все ДЗ у мальчиков: {avg_homework_rating_for_sex(students_list)["male"]}, у девочек {avg_homework_rating_for_sex(students_list)["female"]}')
        print(f'Средняя оценка за экзамены у мальчиков: {avg_exam_rating_for_sex(students_list)["male"]}, у девочек {avg_exam_rating_for_sex(students_list)["female"]}')
    elif command == 3:
        print(f'Средняя оценка за все ДЗ у студентов с опытом программирования: {avg_homework_rating_for_experiece(students_list)["have_exp"]}, без опыта программирования {avg_homework_rating_for_experiece(students_list)["not_have_exp"]}')
        print(f'Средняя оценка за экзамены у студентов с опытом программирования: {avg_exam_rating_for_experiece(students_list)["have_exp"]}, без опыта программирования {avg_exam_rating_for_experiece(students_list)["not_have_exp"]}')    
    else:
        print(f'Вы ввели неизвестную команду "{command}"')
              
              
get_result()
```

```python


# Подулючаем библиотеку
import numpy as np


def all_homework_rating(students_dict, groupby=None):
    
    """
    Возвращает среднюю оценку за все ДЗ по группе (округление до 2 десятичного знака)
    groupby - признак, на основе которого будет происходить агрегирование данных
        Аргумент может принимать значения
         - None - без группировки. Просто среднее значение по группе
         - sex - группировка по полу
         - experience - группировка по наличию опыта в программировании
    """
    

    
    # Если у нас нет группировки
    if not groupby:
        # Здесь храним все оценки за ДЗ
        hw_rating = []

        for x in students_dict:
            for grade in x['grade']:
                hw_rating.append(int(grade))

        return round(np.mean(hw_rating),2)

    
    
    
    # Если нужна группировка по полу
    elif groupby == 'sex':
        # Здесь храним все оценки за ДЗ для каждого пола
        male_homework_rating = []
        female_homework_rating = []

        for x in students_dict:
            if x['sex'] == 'м':            
                for grade in x['grade']:
                    male_homework_rating.append(int(grade))
            else:
                for grade in x['grade']:
                    female_homework_rating.append(int(grade))

        return {'male':round(np.mean(male_homework_rating),2), 'female': round(np.mean(female_homework_rating),2)}
    
    
    
    # Если нужна группировка по наличию опыта в программировании
    elif groupby == 'experience':
        # Здесь храним все оценки за ДЗ для тех, кто имеет и не имеет опыта программирования
        have_prog_exp_rating = []
        not_have_prog_exp_rating = []

        for x in students_dict:
            if x['program_exp']:            
                for grade in x['grade']:
                    have_prog_exp_rating.append(int(grade))
            else:
                for grade in x['grade']:
                    not_have_prog_exp_rating.append(int(grade))

        return {'have_exp': round(np.mean(have_prog_exp_rating),2), 'not_have_exp': round(np.mean(not_have_prog_exp_rating),2)}
    
    else:
        print(f'Недопустимый параметр для группировки "{groupby}".')
        print(f'Выберите один из типов группировки:')
        print(f'\t "groupby=sex" - Группировка по полу')
        print(f'\t "groupby=experience" - Группировка по наличию опыта программирования')











def all_exams_rating(students_dict, groupby=False):
    
    """
    Возвращает среднюю оценку за экзамены по группе (округление до 2 десятичного знака)
    groupby - признак, на основе которого будет происходить агрегирование данных
        Аргумент может принимать значения
         - None - без группировки. Просто среднее значение по группе
         - sex - группировка по полу
         - experience - группировка по наличию опыта в программировании
    """
    

    
    #
    # Если у нас нет группировки
    # Нужно только среднее значение
    #
    if not groupby:
        # Здесь храним все оценки за экзамены
        exam_rating = []

        for x in students_dict:
            exam_rating.append(int(x['exam']))

        return round(np.mean(exam_rating),2)

    
    
    #
    # Если нужна группировка по полу
    #
    elif groupby == 'sex':
        # Здесь храним все оценки за экзамены для каждого пола
        male_exam_rating = []
        female_exam_rating = []

        for x in students_dict:
            if x['sex'] == 'м':            
                    male_exam_rating.append(int(x['exam']))
            else:
                    female_exam_rating.append(int(x['exam']))

        return {'male': round(np.mean(male_exam_rating),2), 'female': round(np.mean(female_exam_rating),2)}
    
    
    #         
    # Если нужна группировка по наличию опыта в программировании
    #
    elif groupby == 'experience':
        # Здесь храним все оценки за экзамены для тех, кто имеет и не имеет опыта программирования
        have_prog_exp_rating = []
        not_have_prog_exprating = []

        for x in students_dict:
            if x['program_exp']:            
                    have_prog_exp_rating.append(int(x['exam']))
            else:
                    not_have_prog_exprating.append(int(x['exam']))

        return {'have_exp': round(np.mean(have_prog_exp_rating),2), 'not_have_exp': round(np.mean(not_have_prog_exprating),2)}
        
        
    # 
    # 
    #
    else:
        print(f'Недопустимый параметр для группировки "{groupby}".')
        print(f'Выберите один из типов группировки:')
        print(f'\t "groupby=sex" - Группировка по полу')
        print(f'\t "groupby=experience" - Группировка по наличию опыта программирования')

        
        
        
        
        
        
def get_result(students_dict):
    # Получаем команду от пользователя (приводим сразу к int)
    command = int(input('Введите команду: '))

    if command == 1:
        print(f'Средняя оценка за все ДЗ по группе: {all_homework_rating(students_dict)}')
        print(f'Средняя оценка за экзамен по группе: {all_exams_rating(students_dict)}')
    elif command == 2:
        print(f'Средняя оценка за все ДЗ у мальчиков: {all_homework_rating(students_dict, groupby="sex")["male"]}, у девочек {all_homework_rating(students_dict, groupby="sex")["female"]}')
        print(f'Средняя оценка за экзамены у мальчиков: {all_exams_rating(students_dict, groupby="sex")["male"]}, у девочек {all_exams_rating(students_dict, groupby="sex")["female"]}')
    elif command == 3:
        print(f'Средняя оценка за все ДЗ у студентов с опытом программирования: {all_homework_rating(students_dict, groupby="experience")["have_exp"]}, без опыта программирования {all_homework_rating(students_dict, groupby="experience")["not_have_exp"]}')
        print(f'Средняя оценка за экзамены у студентов с опытом программирования: {all_exams_rating(students_dict, groupby="experience")["have_exp"]}, без опыта программирования {all_exams_rating(students_dict, groupby="experience")["not_have_exp"]}')
    else:
        print(f'Вы ввели неизвестную команду "{command}"')
              
              

              
              
              
              
# Вызываем основную функцию              
get_result(students_list)
```
