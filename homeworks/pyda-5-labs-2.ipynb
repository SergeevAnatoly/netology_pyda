```python
# 6. Выдвинуть несколько гипотез по набору имеющихся переменных и проверить эти гипотезы с использованием t-test.
# Гипотеза 1: двудверные автомобили стоят всегда дешевле 3+ дверных. Для более верного эксперимента возьмем еще условие, что все авто переднеприводные

two_doors = df.loc[ (df['num-of-doors'] == 'two') & (df['drive-wheels'] == 'fwd')]['price']
other_doors = df.loc[ (df['num-of-doors'] != 'two') & (df['drive-wheels'] == 'fwd')]['price']
```

```python
# Смотрим на вариацию
statistics.variance(two_doors), statistics.variance(other_doors)

# Они сильно отличаются. Делаем через Walch
```

```python
tt = stats.ttest_ind(two_doors, other_doors, equal_var=False)
print(f'\033[1mВЫВОД: Т.к. p-value({round(tt[1],2)})<5%, то мы отвергаем гипотезу.')
```

```python
# Гипотеза 2 `sedan` стоит дороже, чем `hatchback`
sedan = df.loc[ (df['body-style'] == 'sedan')]['price']
hatch = df.loc[ (df['body-style'] == 'hatchback')]['price']

tt2 = stats.ttest_ind(sedan, hatch)

print(f'{tt2}')
print(f'\033[1mВЫВОД: Т.к. p-value({round(tt[1],4)})<5%, то мы отвергаем гипотезу.')
```
