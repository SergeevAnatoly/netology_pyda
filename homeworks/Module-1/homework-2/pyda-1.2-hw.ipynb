## Задание 3

Дан список поисковых запросов. Получить распределение количества слов в них. 
Т. е. поисковых запросов из одного - слова 5%, из двух - 7%, из трех - 3% и т.д.

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
```

```python
# Для упрощения принимаем, что союзы и предлоги тоже являются "словами"

# В этой переменной храним количество слов для каждого запроса
queries_lens = []
for query in queries:
    queries_lens.append(len(query.split(' ')))

# Создаем множество с уникальными значениями
unique_lens = set(queries_lens)

# Пробегаемся по списку с количеством слов в поисковом запросе и производим расчеты
for x in unique_lens:
    print(f'Если слов в поисковом запросе {x}, то их {round(queries_lens.count(x)/len(queries_lens)*100, 2)}%')
```

```python
"""
Если мы хотим исключить разные части речи из расчетов, которые (чаще всего) игнорируются при анализе запроса
(предлоги, союзы и так далее)
"""

# Создаем список элементов, которые не будем учитывать при подсчетах
negatives = ['без', 'в', 'для', 'за', 'из', 'к', 'на', 'над', 'о', 'об', 'от', 'по', 'под', 'пред', 'при', 'про', 'с', 'у', 'через', 'а', 'и', 'но', 'да', 'зато', 'если', 'потому', 'поэтому', 'тоже', 'также']

# Счетчик для цикла while
count = 0

# Список для храрения обработанных запросов (после удаления элементов из списка negatives)
queries_new = []

# Циклом пробегаемся по нашим запросам в списке
while count < len(queries):
    # Разбиваем поисковый запрос на слова по "пробелу"   
    _query = queries[count].split(" ")
    
    # Бежим по списку с исключениями и ищем их в _query
    for negative in negatives:
        
    # Если нашли слово из исключения, то удаляем его
        if negative in _query:
            mini_count = 0
            
            # Подстразовываем себя от ситуации, когда слово-исключение встречается в поисковом запросе более 1 раза
            # Например, 'но я же не могу написать слово но' :)
            while mini_count < len(_query):
                
                # Если нашли слово-исключение, удаляем его по индексу
                if negative == _query[mini_count]:
                    _query.remove(negative)
                    
                # увеличиваем итератор
                mini_count += 1
                
    # Добавляем список со словами из запроса в новый список
    queries_new.append(_query)
    count += 1
    
# В этой переменной храним количество слов для каждого запроса
queries_lens = []
for query in queries_new:
    queries_lens.append(len(query))

# Создаем множество с уникальными значениями
unique_lens = set(queries_lens)

# Пробегаемся по списку с количеством слов в поисковом запросе и производим расчеты
for x in unique_lens:
    print(f'Если слов в поисковом запросе {x}, то их {round(queries_lens.count(x)/len(queries_lens)*100, 2)}%')

```

## Задание 4

Дана статистика рекламных каналов по объемам продаж. Напишите скрипт, который возвращает название канала с максимальным объемом.
Т. е. в данном примере скрипт должен возвращать 'yandex'.

```python
stats = {'facebook': 55, 'yandex': 120, 'vk': 115, 'google': 99, 'email': 42, 'ok': 98}
```

```python
maximum = 0
maximum_search = ''

for key, value in stats.items():
    if value > maximum:
        maximum = value
        maximum_search = key
        
print(f'Максимальный объем продаж у "{maximum_search}"')
```

## Задание 5

Дан поток логов по количеству просмотренных страниц для каждого пользователя. Список отсортирован по ID пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. 
Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]
```

```python
# В этом листе переменной храним ID пользователе
users = []

# В этом листе храним количество просмотров
views = []

# Бежим по изначальному листу
for s in stream:
    # Записываем ID пользователя в отдельный лист
    users.append(s.split(",")[1])
    
    # Записываем количество просмотров в отдельный лист
    views.append(int(s.split(",")[2]))
    
# Производим высчисления с округлением (суммируем количество просморов и делим на количество уникальных ID пользователей)
print(f'Среднее значение просмотров на пользователя равно {round(sum(views) / len(set(users)),2)}')
```

## Задание 6

Дана статистика рекламных кампаний по дням. Напишите алгоритм, который по паре дата-кампания ищет значение численного столбца. 
Т. е. для даты '2018-01-01' и 'google' нужно получить число 25. 
Считайте, что все комбинации дата-кампания уникальны.

```python
stats = [
    ['2018-01-01', 'google', 25],
    ['2018-01-01', 'yandex', 65],
    ['2018-01-01', 'market', 89],
    ['2018-01-02', 'google', 574],
    ['2018-01-02', 'yandex', 249],
    ['2018-01-02', 'market', 994],
    ['2018-01-03', 'google', 1843],
    ['2018-01-03', 'yandex', 1327],
    ['2018-01-03', 'market', 1764],
]
```

```python
# Получаем данные от пользователя (корректность не проверяем). Переводим к типу "строка"
date = str(input('Введите дату: '))
source = str(input('Введите источник: '))
```

```python
# В этом словаре будем хранить данные для поиска
stats_dict = {}

# Бежим по изначальному списку
for element in stats:
    # Создаем словарь в парой "ключ:значение" вида {'2018-01-03:google' : 1843}
    stats_dict.update({str(element[0]) + ':' + str(element[1]) : int(element[2])})

# Достаем значение из собранного словаря (предварительно собрав ключ на основе данных, введенных пользователем)
# Создаем ключ для поиска из даты и источника, которые ввел пользователь. Получаем ключ вида '2018-01-03:market'
# И выводим данные на экран
print(f'Количество посещений равно {stats_dict.get(date + ":" + source)}')
```
