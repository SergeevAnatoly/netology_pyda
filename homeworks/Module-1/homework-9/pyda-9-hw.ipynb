## Задача 1

Напишите функцию date_range, которая возвращает список дней между датами _start_date_ и _end_date_. Даты должны вводиться в формате `YYYY-MM-DD`.

```python
from datetime import datetime, timedelta
import traceback
```

```python
def date_range(start_date, end_date):
    """
    Функция возвращает список дней между датами start_date и end_date.
    Даты должны вводиться в формате `YYYY-MM-DD`.
    """
    
    # Список для хранения результата
    date_range = []
    
    # Оборачиваем все в глобальный try...except
    try:
        
        # Для поимки ошибки при вводе даты НАЧАЛА периода оборачиваем приведение к типу datetime try...except
        try:
            _start_date = datetime.strptime(start_date, "%Y-%m-%d")
        except ValueError:
            print(f'Ошибка при обработке входящих данных:\n\tНеверный формат ввходных данных в параметре `start_date`\n\tДата должна вводиться в формате `YYYY-MM-DD`')
            
            
        # Для поимки ошибки при вводе даты КОНЦА периода оборачиваем приведение к типу datetime try...except
        try:
            _end_date = datetime.strptime(end_date, "%Y-%m-%d")
        except ValueError:
            print(f'Ошибка при обработке входящих данных:\n\tНеверный формат ввходных данных в параметре `end_date`\n\tДата должна вводиться в формате `YYYY-MM-DD`')
        
        
        # Тут храним текущую дату в цикле (в начале она равна дате начала периода)
        current_date = _start_date
        
        # крутим цикл, пока текущая дата не более даты КОНЦА периода
        while current_date <= _end_date:
            
            # В результирующий список записываем результат смещения
            date_range.append(current_date.strftime('%Y-%m-%d'))
            
            # Смещаем текущую дату на 1 день ВПЕРЕД
            current_date += timedelta(days=1)
            
        
        # Возвращаем результат
        return date_range

    except Exception:
        print(traceback.print_exc())
        

# Вызываем исходную функцию
date_range('2019-01-11', '2019-11-12')
    
```

В последнем примере поиска по словарю мы использовали 3 столбца. Напишите функцию, которая формирует словарь для поиска по n столбцам.

```python
def find_in_dict_with_n_cols_autos(file):
    """
    Функция формирует словарь для поиска по n столбцам
    Она читает файл, смотрит на количество столбцов и создает словарь по n-1 колонкам
    """
    stats_dict = {}
    num_cols = 0
    
    try:
    
        with open(f'{file}') as f:
            for line in f:
                num_cols = len(line.strip().split(',')) - 1
                break



        with open(f'{file}') as f:
            for line in f:
                line = line.strip().split(',')
                stats_dict[tuple(line[:num_cols])] = line[num_cols]
                
    except:
        print('Ошибка обработки данных')


    return stats_dict
    
```
